```python
import matplotlib.pyplot as plt
import numpy as np

def read_single_column_text_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(float(line.strip()))  # Assuming the values are floating-point numbers
    return data

def plot_two_single_column_text_files(file_path1, file_path2):
    data1 = read_single_column_text_file(file_path1)
    data2 = read_single_column_text_file(file_path2)

    plt.plot(data1, label='Source IP')
    plt.plot(data2, label='Destination IP')
    #plt.xlabel('Time (1 Sec)')
    plt.title('Weibull Shape Parameter')
    #plt.ylabel('Weibull Shape Parameter')
    #plt.title('Flow Length Distribution')
    plt.legend()
    #plt.grid(True)
    #plt.show()

def plot_two_single_column_text_files_2(file_path1, file_path2):
    data1 = read_single_column_text_file(file_path1)
    data2 = read_single_column_text_file(file_path2)

    plt.plot(data1, label='Source IP')
    plt.plot(data2, label='Destination IP')
    plt.xlabel('Observation Time (1 Sec)')
    plt.title('Weibull Scale Parameter')
    #plt.ylabel('Weibull Shape Parameter')
    #plt.title('Flow Length Distribution')
    plt.legend()
    #plt.grid(True)
    #plt.show()
def plot_two_single_column_text_files_F0(file_path1, file_path2):
    data1 = read_single_column_text_file(file_path1)
    data2 = read_single_column_text_file(file_path2)

    plt.plot(data1, label='Source IP')
    plt.plot(data2, label='Destination IP')
    plt.xlabel('Observation Time (1 Sec)')
    plt.title('F0')
    #plt.ylabel('Weibull Shape Parameter')
    #plt.title('Flow Length Distribution')
    plt.legend()
    #plt.grid(True)
    #plt.show()

def plot_two_single_column_text_files_F1(file_path1, file_path2):
    data1 = read_single_column_text_file(file_path1)
    data2 = read_single_column_text_file(file_path2)
    plt.plot(data1, label='Source IP')
    plt.plot(data2, label='Destination IP')
    plt.xlabel('Observation Time (1 Sec)')
    plt.title('F1')
    #plt.ylabel('Weibull Shape Parameter')
    #plt.title('Flow Length Distribution')
    plt.legend()
    #plt.grid(True)
    #plt.show()

def plot_two_single_column_text_files_F2(file_path1, file_path2):
    data1 = read_single_column_text_file(file_path1)
    data2 = read_single_column_text_file(file_path2)

    plt.plot(data1, label='Source IP')
    plt.plot(data2, label='Destination IP')
    plt.xlabel('Observation Time (1 Sec)')
    plt.title('F2')
    #plt.ylabel('Weibull Shape Parameter')
    #plt.title('Flow Length Distribution')
    plt.legend()
    #plt.grid(True)
    #plt.show()
def calculate_and_plot_FD(file_path1, file_path2, file_path3, file_path4, time):
    src_ip_f0 = read_single_column_text_file(file_path1)
    src_ip_f2 = read_single_column_text_file(file_path2)

    dest_ip_f0 = read_single_column_text_file(file_path3)
    dest_ip_f2 = read_single_column_text_file(file_path4)

    absolute_times = read_single_column_text_file(time)

    src_ip_f02np = np.array(src_ip_f0)
    src_ip_f22np = np.array(src_ip_f2)
    dest_ip_f02np = np.array(dest_ip_f0)
    dest_ip_f22np = np.array(dest_ip_f2)

    FD = (src_ip_f22np/src_ip_f02np) / (dest_ip_f22np/dest_ip_f02np)
    plt.figure(figsize=(10, 4))
    plt.plot(absolute_times,FD, label='FD',marker=".")
    plt.xlabel('Time (s)')
    plt.ylabel('FD Value')
    plt.legend()
    
    plt.show()
    
def calculate_F2_normalize_and_plot_FD(file_path1, file_path2, file_path3, file_path4, time):
    src_ip_f1_values = read_single_column_text_file(file_path1)
    src_ip_f2_values = read_single_column_text_file(file_path2)

    dest_ip_f1_values = read_single_column_text_file(file_path3)
    dest_ip_f2_values = read_single_column_text_file(file_path4)

    absolute_times = read_single_column_text_file(time)

    src_ip_f1_values2np = np.array(src_ip_f1_values)
    src_ip_f2_values2np = np.array(src_ip_f2_values)

    dest_ip_f1_values2np = np.array(dest_ip_f1_values)
    dest_ip_f2_values2np = np.array(dest_ip_f2_values)
    plt.figure(figsize=(10, 4))
    plt.plot(absolute_times, (src_ip_f2_values2np/(src_ip_f1_values2np**2)), label='F2 Normalize Values of Source IP',marker=".")
    plt.plot(absolute_times, (dest_ip_f2_values2np/(dest_ip_f1_values2np**2)), label='F2 Normalize Values of Destination IP',marker=".")
    
    plt.xlabel('Time (s)')
    plt.ylabel('Count')
    plt.title('F2 of Src & Dest IP Count over Time')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    #shape
    file_path2 = "dest_ip_weibull_k.txt"  # Replace this with the path to your file
    file_path1 = "src_ip_weibull_k.txt"  # Replace this with the path to your file
    #plot_two_single_column_text_files(file_path1, file_path2)
    #scale
    file_path4 = "dest_ip_weibull_theta.txt"  # Replace this with the path to your file
    file_path3 = "src_ip_weibull_theta.txt"  # Replace this with the path to your file
    #plot_two_single_column_text_files_2(file_path3, file_path4)
    #plt.show()

    file_path5 = "src_ip_F0_values.txt"  # Replace this with the path to your file
    file_path6 = "dest_ip_F0_values.txt"  # Replace this with the path to your file
    file_path7 = "src_ip_F1_values.txt"  # Replace this with the path to your file
    file_path8 = "dest_ip_F1_values.txt"  # Replace this with the path to your file
    file_path9 = "src_ip_F2_values.txt"  # Replace this with the path to your file
    file_path10 = "dest_ip_F2_values.txt"  # Replace this with the path to your file
    file_path11 = "absolute_time.txt"

# First subplot
# plt.subplot(2, 1, 1)  # 2 row, 1 columns, index 1 (upper subplot)
# plot_two_single_column_text_files(file_path1, file_path2)
# # Second subplot
# plt.subplot(2, 1, 2)  # 1 row, 2 columns, index 2 (lower subplot)
# plot_two_single_column_text_files_2(file_path3, file_path4)
# # Adjust layout to prevent overlap
# plt.tight_layout()
# plt.show()

# plot_two_single_column_text_files_F0(file_path5, file_path6)
# plt.show()
# plot_two_single_column_text_files_F1(file_path7, file_path8)
# plt.show()
# plot_two_single_column_text_files_F2(file_path9, file_path10)
# plt.show()
```
